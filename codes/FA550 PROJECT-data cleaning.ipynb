```python
language_counts = {
    'English': 4866102,
    'Spanish': 69472,
    'German': 5099,
    'French': 4318,
    'Portuguese': 2356,
    'Italian': 2199,
    'Afrikaans': 1531,
    'Chinese': 475,
    'Russian': 423,
    'Japanese': 421,
    'Dutch': 400,
    'Catalan': 64,
    'Unknown': 56,
    'Swedish': 52,
    'Norwegian': 43,
    'Irish': 41,
    'Danish': 37,
    'Hindi': 31,
    'Finnish': 21,
    'Icelandic': 12,
    'Romansh': 11,
    'Tamil': 9,
    'Bokml Norwegian': 5,
    'Castilian': 4,
    'Multilingual': 3,
    'Galician': 3,
    'Welsh': 2,
    'Provenal': 1
}

selected_reviews = pd.DataFrame()

max_reviews_to_sample = 5000 

for language, count in language_counts.items():
    # Calculate 80% of the total occurrences for each language
    print('Language:', language)
    if language in ['English', 'Spanish']:
        num_rows_to_select = np.random.randint(1000,min(count, max_reviews_to_sample))
        print('num_rows_to_select', num_rows_to_select)
    else:
        # For other languages, sample up to 80% of the occurrences
        eighty_percent = int(0.8 * count)
        num_rows_to_select = 1 if eighty_percent < 1 else np.random.randint(1, min(count, eighty_percent) + 1)
        print('num_rows_to_select', num_rows_to_select)

    # Select random rows for the current language
    language_reviews = filtered_data_common_asin[filtered_data_common_asin['language'] == language].sample(n=num_rows_to_select, random_state=42)
    
    selected_reviews = pd.concat([selected_reviews, language_reviews])
```
